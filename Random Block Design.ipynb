```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api  import ols
```

```python
ex = pd.read_excel(r'C:\Users\tyagir01\Desktop\Stress.xlsx')
ex
```

```python
df = pd.melt(ex.reset_index(), id_vars=['index'], value_vars=['System A', 'System B', 'System C'])
df.columns = ['index', 'treatments', 'value']
df
```

```python
model = ols('value ~ C(treatments)',data = df).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

```python
df1 = pd.melt(ex.reset_index(),id_vars=['index'], value_vars=['System A', 'System B', 'System C'])
df1.columns = ['blocks','treatments', 'value']
df1
```

```python
model = ols('value ~ C(blocks) + C(treatments)', data = df1).fit()
new_anova_table = sm.stats.anova_lm(model, typ=1)
new_anova_table
```
