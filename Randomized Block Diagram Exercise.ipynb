```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.read_excel(r'C:\Users\tyagir01\Desktop\Data Analytics Problem Set\Chem.xlsx')
df
```

```python
df1 = pd.melt(df.reset_index(), id_vars = 'index', value_vars = ['Chem 1','Chem 2','Chem 3','Chem 4'])
df1.columns = ['index', 'treatments','value']
df1
```

```python
model = ols('value~C(treatments)', data=df1).fit()
aov_table = sm.stats.anova_lm(model, typ=1)
aov_table
```

```python
df2 = pd.melt(df.reset_index(), id_vars = 'index', value_vars = ['Chem 1', 'Chem 2', 'Chem 3', 'Chem 4'])
df2.columns = ['Fabric', 'Chemical', 'value' ]
df2
```

```python
model = ols('value ~ C(Fabric) + C(Chemical)', data = df2).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```
