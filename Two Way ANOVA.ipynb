```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
fn = 'Value ~ C(College) + C(Prep_pro) + C(College):C(Prep_pro)'
model = ols(fn, df).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```
